# eRisk T2 dataset browser

Ноутбук для быстрого просмотра пользователей, их лейблов и отдельных тредов.
Датасет читается напрямую из `Downloads`, без копирования в проект.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'experiments').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.append(str(PROJECT_ROOT / 'experiments'))

from erisk_t2_browser import (
    DEFAULT_DATASET_DIR,
    label_name,
    list_users,
    render_thread,
    summarize_user,
    user_activity_table,
)

DATASET_DIR = DEFAULT_DATASET_DIR
DATASET_DIR
```

```python
users = list_users(DATASET_DIR)
len(users), users[0]
```

```python
sample_rows = [
    {
        'user_id': user.user_id,
        'label': label_name(user.label),
        'file_size_mb': user.file_size_mb,
    }
    for user in users[:20]
]
sample_rows
```

```python
positive_users = [u for u in users if u.label == 1]
control_users = [u for u in users if u.label == 0]
{
    'total_users': len(users),
    'positive_users': len(positive_users),
    'control_users': len(control_users),
    'largest_files_mb': sorted((u.file_size_mb for u in users), reverse=True)[:10],
}
```

## Выбери пользователя

Меняй `USER_ID`, затем перезапускай ячейки ниже.
Удобно брать id из `sample_rows` или из списков `positive_users` / `control_users`.

```python
USER_ID = positive_users[0].user_id
USER_ID
```

```python
summary = summarize_user(USER_ID, DATASET_DIR)
summary['label'] = label_name(next(u.label for u in users if u.user_id == USER_ID))
summary
```

```python
activity = user_activity_table(USER_ID, DATASET_DIR, target_only=True, max_rows=50)
activity[:20]
```

## Посмотреть конкретный тред

Бери `thread_index` из `activity` и открывай полный тред ниже.

```python
THREAD_INDEX = activity[0]['thread_index']
THREAD_INDEX
```

```python
print(render_thread(USER_ID, THREAD_INDEX, DATASET_DIR))
```

## pandas DataFrame по пользователю

Собирает все target-сообщения выбранного `USER_ID` в `pd.DataFrame`.

```python
import pandas as pd
```

```python
user_df = pd.DataFrame(user_activity_table(USER_ID, DATASET_DIR, target_only=True, max_rows=None))
user_df
```

```python
user_df.sort_values(['created_utc', 'thread_index']).reset_index(drop=True)
```

```python
all_rows = []

for user in users:
    rows = user_activity_table(user.user_id, DATASET_DIR, target_only=True, max_rows=None)
    for row in rows:
        row['user_id'] = user.user_id
        row['label'] = label_name(user.label)
    all_rows.extend(rows)

all_df = pd.DataFrame(all_rows)
all_df.head()
```

```python
all_df.groupby('user_id').size().sort_values(ascending=False).head(20)
```

```python
all_df.groupby(['user_id', 'kind']).size().unstack(fill_value=0).head(20)
```
